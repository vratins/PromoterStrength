```python
def get_longest_length(sequences):
    longest_length = 0
    for sequence in sequences:
        if len(sequence) > longest_length:
            longest_length = len(sequence)
    return longest_length
```

```python
from transformers import AutoTokenizer, AutoModelForMaskedLM
import torch

# Import the tokenizer and the model
tokenizer = AutoTokenizer.from_pretrained("InstaDeepAI/nucleotide-transformer-v2-50m-multi-species")
model = AutoModelForMaskedLM.from_pretrained("InstaDeepAI/nucleotide-transformer-v2-50m-multi-species")

# Choose the length to which the input sequences are padded. By default, the
# model max length is chosen, but feel free to decrease it as the time taken to
# obtain the embeddings increases significantly with it.
print('len')
 #tokenizer.model_max_length

# Create a dummy dna sequence and tokenize it
sequences =df['seq'][:10].tolist()#['AAGGGCCGTCGGTGATTTTAGTCTTCCTCAGTGGTTACCAGGTCTCCCGTTCTCTCCACATTCTCTTCGCACTCGATCGGAGCGTAAGGTCCGTGCCAAGTCGCCGAGCGCTTTCGAGTGCCGTCTCTGGGTAGGGT'] #df['seq'].tolist()#["AAGGGCCGTCGGTGATTTTAGTCTTCCTCAGTGGTTACCAGGTCTCCCGTTCTCTCCACATTCTCTTCGCACTCGATCGGAGCGTAAGGTCCGTGCCAAGTCGCCGAGCGCTTTCGAGTGCCGTCTCTGGGTAGGGT", "ATTTCTCTCTCTCTCTGAGATCGATCGATCGAT"]
max_length = get_longest_length(sequences)
print(f"longest length: {max_length}")
tokens_ids = tokenizer.batch_encode_plus(sequences, return_tensors="pt", padding="max_length", max_length = max_length)["input_ids"]
print('computing')
# Compute the embeddings
attention_mask = tokens_ids != tokenizer.pad_token_id
torch_outs = model(
    tokens_ids,
    attention_mask=attention_mask,
    encoder_attention_mask=attention_mask,
    output_hidden_states=True
)
print('out')
# Compute sequences embeddings
embeddings = torch_outs['hidden_states'][-1].detach().numpy()
print(f"Embeddings shape: {embeddings.shape}")
print(f"Embeddings per token: {embeddings}")

# Add embed dimension axis
attention_mask = torch.unsqueeze(attention_mask, dim=-1)

# Compute mean embeddings per sequence
mean_sequence_embeddings = torch.sum(attention_mask*embeddings, axis=-2)/torch.sum(attention_mask, axis=1)
print(f"Mean sequence embeddings: {mean_sequence_embeddings}")
print(f"Mean sequence embeddings shape: {mean_sequence_embeddings.shape}")
```
